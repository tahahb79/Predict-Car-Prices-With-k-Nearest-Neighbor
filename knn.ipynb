```python
import os
import tensorflow as tf
import keras
import sklearn
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
```

```python
ROOt="/content/dataset/car/car.data"
data = pd.read_csv(ROOt)
print (data.head())

```

```python
le = preprocessing.LabelEncoder()
buying =le.fit_transform(list(data["buying"]))
maint=le.fit_transform(list(data["maint"]))
doors=le.fit_transform(list(data["doors"]))
persons=le.fit_transform(list(data["persons"]))
lug_boot=le.fit_transform(list(data["lug_boot"]))
safety=le.fit_transform(list(data["safety"]))
classes=le.fit_transform(list(data["class"]))
print(classes,safety)
```

```python
x = list(zip(buying,maint,doors,persons,lug_boot,safety))
y= list(classes)

x_train,x_test,y_train,y_test=sklearn.model_selection.train_test_split(x,y,test_size=0.2)

# print (y_train)

model =KNeighborsClassifier(n_neighbors=9)
model_fit = model.fit(x_train,y_train)
acc = model_fit.score(x_test,y_test)
print(acc)

predicted = model_fit.predict(x_test)
cls_string = ["unacc","acc","good","vgood"]
for x in range (len(predicted)):
  print("pred: ",cls_string[predicted[x]],"real: ",cls_string[y_test[x]] )


```
